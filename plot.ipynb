```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import matplotlib.cm as cm
import re  # Import thư viện để sử dụng regular expressions
import numpy as np  # Để tính trung bình

# Đường dẫn tới folder chứa các thư mục bs_alpha
folder_path = "D:\\CS331\\GAN\\Model\\Test"  # Thay thế bằng đường dẫn thực tế tới folder

# Tạo các danh sách để lưu trữ dữ liệu cho từng loại biểu đồ
epochs = []
loss_values = []
psnr_values = []
ssim_values = []
labels = []  # Danh sách để lưu tên phiên bản
alphas = []  # Danh sách để lưu alpha từ tên thư mục

# Duyệt qua các thư mục con trong folder
for subfolder in os.listdir(folder_path):
    subfolder_path = os.path.join(folder_path, subfolder)
    
    # Kiểm tra xem đó có phải là thư mục và tên folder bắt đầu bằng 'bs'
    if os.path.isdir(subfolder_path) and subfolder.startswith('bs'):
        
        # Sử dụng regular expression để tìm alpha trong tên thư mục
        match = re.search(r'bs(\d+)_alpha(\d+)', subfolder)  # Tìm 'bsXX_alphaYY'
        if match:
            bs_value = int(match.group(1))  # Lấy giá trị bs (XX)
            alpha_value = int(match.group(2))  # Lấy giá trị alpha (YY)
            alphas.append(alpha_value)
        else:
            alphas.append(None)  # Nếu không tìm thấy alpha, gán giá trị None
        
        # Duyệt qua các file trong thư mục con
        for filename in os.listdir(subfolder_path):
            if filename.endswith("gridsearch_log.txt"):
                file_path = os.path.join(subfolder_path, filename)

                # Đọc dữ liệu từ file .txt
                data = pd.read_csv(file_path, sep=",\s*", engine="python")

                # Lưu dữ liệu vào các danh sách
                epochs.append(data["Epoch"])
                loss_values.append(data["Avg L1 Loss"])
                psnr_values.append(data["Avg PSNR"])
                ssim_values.append(data["Avg SSIM"])
                labels.append(subfolder)  # Lưu tên thư mục (phiên bản)

# Tính trung bình theo alpha
def average_by_alpha(values, alphas):
    # Tạo dictionary để nhóm các giá trị theo alpha
    alpha_grouped = {}
    for i, alpha in enumerate(alphas):
        if alpha not in alpha_grouped:
            alpha_grouped[alpha] = []
        alpha_grouped[alpha].append(values[i])

    # Tính trung bình cho mỗi nhóm alpha
    alpha_avg = {alpha: np.mean(alpha_grouped[alpha], axis=0) for alpha in alpha_grouped}
    return alpha_avg

# Tính trung bình cho các giá trị Loss, PSNR, SSIM theo alpha
average_loss = average_by_alpha(loss_values, alphas)
average_psnr = average_by_alpha(psnr_values, alphas)
average_ssim = average_by_alpha(ssim_values, alphas)

# Tạo bảng màu (colormap) để đảm bảo các biểu đồ có màu đồng nhất
unique_alphas = sorted(set(alphas))  # Các giá trị alpha duy nhất
colors = cm.get_cmap('tab10', len(unique_alphas))  # Sử dụng bảng màu 'tab10' để phân biệt các alpha

# Tạo dictionary ánh xạ alpha với màu sắc
alpha_color_map = {alpha: colors(i) for i, alpha in enumerate(unique_alphas)}

# Vẽ biểu đồ Loss (trung bình theo alpha)
plt.figure(figsize=(10, 6))
for alpha in unique_alphas:
    plt.plot(epochs[0], average_loss[alpha], label=f"Loss - Alpha {alpha}", marker="o", color=alpha_color_map[alpha])
plt.xlabel("Epoch")
plt.ylabel("Avg L1 Loss")
plt.title("Loss Comparison Across Alpha")
plt.legend()  # Đặt legend ở góc trên bên phải
plt.grid(True)
plt.show()

# Vẽ biểu đồ PSNR (trung bình theo alpha)
plt.figure(figsize=(10, 6))
for alpha in unique_alphas:
    plt.plot(epochs[0], average_psnr[alpha], label=f"PSNR - Alpha {alpha}", marker="o", color=alpha_color_map[alpha])
plt.xlabel("Epoch")
plt.ylabel("Avg PSNR")
plt.title("PSNR Comparison Across Alpha")
plt.legend()  # Đặt legend ở góc trên bên phải
plt.grid(True)
plt.show()

# Vẽ biểu đồ SSIM (trung bình theo alpha)
plt.figure(figsize=(10, 6))
for alpha in unique_alphas:
    plt.plot(epochs[0], average_ssim[alpha], label=f"SSIM - Alpha {alpha}", marker="x", color=alpha_color_map[alpha])
plt.xlabel("Epoch")
plt.ylabel("Avg SSIM")
plt.title("SSIM Comparison Across Alpha")
plt.legend()  # Đặt legend ở góc trên bên phải
plt.grid(True)
plt.show()
```
